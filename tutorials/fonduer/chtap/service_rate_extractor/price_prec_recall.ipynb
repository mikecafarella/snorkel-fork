```python
xls_label_file = pd.ExcelFile('Labeling_Spreadsheet_Texas_Profiles_Nov_2017.xlsx')
df_labeled = xls_label_file.parse('sheet1')

df_labeled = df_labeled.apply(lambda x: x.astype(str).str.lower())
result=  pd.merge(df_labeled, df_pred, on='file name')
result
```

```python
"""Change data frame column names"""# df.rename(index=str, columns={"A": "a", "B": "c"})
result = result.rename(index=str,columns = {"incall price [$XX.XX, $XX.XX...]":"hand_labeled_incall_price",
                                 "incall time unit [# of hours,#of hours2...], 12 for overnight":"incall_time_unit",
                                 'outcall price [$XX.XX]':"hand_labeled_outcall_price",
                                            "outcall time unit [# of hours], 12 for overnight":"outcall_time_unit"})

price_df_orig = result[["file name","hand_labeled_incall_price","incall_time_unit","hand_labeled_outcall_price","outcall_time_unit","price_pred"]]
price_df_orig
```

```python
''' remove $ from hand labeled data'''
import numpy

price_df= price_df_orig.replace({'\$': '', ' ': ''}, regex=True)

def convert_to_int(row):
    x = row.split(",")
    lst=[]
    for i in x:
        if i!="none":
            y = float(i)
            lst.append(y)
    
    
    return lst
price_df['hand_labeled_incall_price'] = price_df['hand_labeled_incall_price'].apply(convert_to_int)

price_df['incall_time_unit'] = price_df['incall_time_unit'].apply(convert_to_int)
price_df['hand_labeled_outcall_price'] = price_df['hand_labeled_outcall_price'].apply(convert_to_int)
price_df['outcall_time_unit'] = price_df['outcall_time_unit'].apply(convert_to_int)

price_df['price_pred'] = price_df['price_pred'].apply(convert_to_int)
price_df
```

```python
uu = [1,2,7,90]
pp= [10,10,12,1]
uu[pp.index(1)]
yy = [1,2,3,np.nan]
#print(np.count_nonzero(~np.isnan(yy)))
import numpy as np
def find_per_hour(row):
    
    x = np.nan
    y = np.nan
    
    if 1 in row["incall_time_unit"]:
        x = row['hand_labeled_incall_price'][row["incall_time_unit"].index(1)]
    if 1 in row["outcall_time_unit"]:
        y =  row['hand_labeled_outcall_price'][row["outcall_time_unit"].index(1)]
    return([x,y])
#print(price_df.ix[0])
find_per_hour(price_df.ix[0])
price_df["labeled_price_per_hour"]= price_df.apply(find_per_hour,axis=1)
price_df
```

```python
'''precision and recal functions'''
import numpy as np
from __future__ import division
import numpy as np
from __future__ import division
what_to_exclude = np.asarray([np.nan]) ## what to exclude from the evaluation

def comp_precision(rr):
    p_hand_labeled = 'labeled_price_per_hour'
    p_pred = 'price_pred'
    
    idx1 = np.setdiff1d(rr[p_hand_labeled],what_to_exclude)

    idx2= rr[p_pred]
    
    idx2 = np.setdiff1d(idx2,what_to_exclude)
    
    nn=np.intersect1d(idx1,idx2)
    
    n=len(np.intersect1d(idx1,idx2))
    m=len(idx2)
    
    if m ==0:
        p = 0
    else:
        p=n/m
    return p


def comp_recall(rr):
    p_hand_labeled = 'labeled_price_per_hour'
    p_pred = 'price_pred'
    
    idx1 = np.setdiff1d(rr[p_hand_labeled],what_to_exclude)
    idx2= rr[p_pred]
    
    idx2 = np.setdiff1d(idx2,what_to_exclude)
    
    n=len(np.intersect1d(idx1,idx2))
    
    m= np.count_nonzero(~np.isnan(idx1))
    
    if m ==0:
        p = 0
    else:
        p=n/m
    
    return p
    
    
```

```python
'''apply the prec and recall on result_selected df'''
import math as math
df_precision = price_df.apply(comp_precision,axis=1)
df_recall = price_df.apply(comp_recall,axis=1)

print "precision:", df_precision.mean()
print "recall:", df_recall.mean()

```
